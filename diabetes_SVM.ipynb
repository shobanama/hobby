```
from sklearn.datasets import load_diabetes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

from sklearn.svm import SVR
```

```
diabetes = load_diabetes()

# Assign X and Y
X = diabetes.data
y = diabetes.target
```

```
diabetes.feature_names
```

```
print("Number of nan's in dataset:",np.sum(np.isnan(X)))
```

```
print('type of x data:', type(X))
print('shape of x data:', X.shape)
print('-'*50)
print('type of y data:', type(y))
print('shape of y data:', y.shape)
print('-'*50)
print('average of y data:', np.mean(y))
print('standard deviation of y_data:', np.std(y))
```

```
#plot each pair (column,target)
for i in range(X.shape[1]):
    print("Feature %d" % (i))
    plt.plot(X[:,i],y,'x')

    plt.ylabel("Target")
    plt.grid()
    plt.show()
```

1)Cross validation (without splitting to test and train data): obtain cross validation score. Model your data using Support Vector Machine(SVM).

```
# SVR

svr_model = SVR()
svr_model.fit(X, y)

score = cross_val_score(svr_model, X, y, cv=10, scoring = 'neg_mean_squared_error')
rmse_scores = np.sqrt(-score)

print("Mean SVR MSE:", -np.mean(score),np.std(score))
print("Mean RSME Score:", (np.mean(rmse_scores)))
```

```
# Split the data
X_train = X[:-30]
X_test = X[-30:]

# Split the targets
y_train = y[:-30]
y_test = y[-30:]

svrMod = SVR()
svrMod.fit(X_train,y_train)
predSvr = svrMod.predict(X_test)

mse = mean_squared_error(y_test,predSvr)
r2 = r2_score(y_test,predSvr)


print('MSE: ',mse)
print('R2: ',r2)
```
